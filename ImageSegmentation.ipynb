# Basics of Image Segmentation 

import PIL.Image as im
import numpy as np

```python
pic = im.open("images/faab2561.jpg").convert("L")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.imshow(np.array(pic), cmap="gray")
```

```python
#mean thresholding
def meanThresh(image):
    return image>np.mean(image)
    
plt.imshow(meanThresh(pic), cmap="gray")
```

```python
#median thresholding
def medianThresh(image):
    return image>np.median(image)
plt.imshow(medianThresh(pic), cmap="gray")
```

```python
plt.hist(np.array(pic).flatten())
```

```python
#Image smoothening
def smoothen(image):
    image = np.array(image)
    for i in range(1, image.shape[0]-1):
        for j in range(1, image.shape[1]-1):
            image[i][j] = (np.sum(image[i-1][j-1:j+2])+np.sum(image[i+1][j-1:j+2]) + image[i][j-1]+image[i][j+1])/9
    return image
        
        
```

```python
smoothPic = smoothen(pic)
```

```python
plt.imshow(smoothPic)
```
